```python
import pandas as pd
import numpy
```

```python
a_data = pd.read_csv('datasets/Agriculture Dataset.csv')
a_data.head()
```

```python
n_data = pd.read_csv('datasets/Food Insecurity.csv')
n_data
```

```python
n_data.drop(n_data.columns[4:6], axis=1, inplace=True)
n_data = n_data.dropna()
n_data["Year"] = n_data["Year"].replace({'2000-2002': '2002'})
n_data["Year"] = n_data["Year"].replace({'2001-2003': '2003'})
n_data["Year"] = n_data["Year"].replace({'2002-2004': '2004'})
n_data["Year"] = n_data["Year"].replace({'2003-2005': '2005'})
n_data["Year"] = n_data["Year"].replace({'2004-2006': '2006'})
n_data["Year"] = n_data["Year"].replace({'2005-2007': '2007'})
n_data["Year"] = n_data["Year"].replace({'2006-2008': '2008'})
n_data["Year"] = n_data["Year"].replace({'2007-2009': '2009'})
n_data["Year"] = n_data["Year"].replace({'2008-2010': '2010'})
n_data["Year"] = n_data["Year"].replace({'2009-2011': '2011'})
n_data["Year"] = n_data["Year"].replace({'2010-2012': '2012'})
n_data["Year"] = n_data["Year"].replace({'2011-2013': '2013'})
n_data["Year"] = n_data["Year"].replace({'2012-2014': '2014'})
n_data["Year"] = n_data["Year"].replace({'2013-2015': '2015'})
n_data["Year"] = n_data["Year"].replace({'2014-2016': '2016'})
n_data["Year"] = n_data["Year"].replace({'2015-2017': '2017'})
n_data["Year"] = n_data["Year"].replace({'2016-2018': '2018'})
n_data["Year"] = n_data["Year"].replace({'2017-2019': '2019'})
n_data["Year"] = n_data["Year"].replace({'2018-2020': '2020'})
n_data["Year"] = pd.to_numeric(n_data["Year"])

n_data
```

```python
exp_data = pd.read_csv('datasets/Expenditure Loans and Investment into Agriculture.csv')
exp_data.head()
```

```python
merged1 = pd.merge(n_data, exp_data, how='inner', left_on=['Area Code','Year'], right_on=['Area Code','Year'])
merged1.head()
```

```python
merged2 = pd.merge(merged1, a_data, how='inner', left_on=['Area Code','Year'], right_on=['Area Code','Year'])
merged2.head()
```

```python
merged = merged2.drop(['Unnamed: 0_y', 'Area_y'], 1)
merged
```

```python
merged.to_csv('nourishment_data.csv')
```
